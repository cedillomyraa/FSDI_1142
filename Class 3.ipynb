```python
#linked list: elements in a list are called "node" the first is the head the last is tail 

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
```

```python
head = Node(0)

new_node1 = Node(1)
new_node2 = Node(2)
new_node3 = Node(3)

#link the nodes above like: 
head.next = new_node1
new_node1.next =new_node2
new_node2.next = new_node3

current = head
while current:
    print(current.data)
    current = current.next
```
